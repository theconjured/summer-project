```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sn_compare = pd.read_csv('./stuff/scale_neighbors_compare.csv')
```

```python
sn_compare
```

```python
scale_varied = sn_compare.head(8)
scale_varied
```

```python
neighbors_varied = sn_compare.tail(8)
neighbors_varied
```

```python
ax = scale_varied.plot(x = 'Scale', y = 'Runtime', style = 'o-', title = 'Scale vs. Runtime')
ax.set_xlabel('Scale')
ax.set_ylabel('Runtime (s)')
plt.show()
```

```python
ax = scale_varied.plot(x = 'Scale', y = ['Guessed', 'False Positive', 'Missed','Likely Accurate'], style = 'o-', title = 'Scale vs. Boxes Found')
ax.set_xlabel('Scale')
ax.set_ylabel('Number of Boxes')
plt.show()
```

```python
ax = neighbors_varied.plot(x = 'Neighbors', y = 'Runtime', style = 'o-', title = 'Neighbors vs. Runtime')
ax.set_xlabel('Neighbors')
ax.set_ylabel('Runtime (s)')
plt.show()
```

```python
ax = neighbors_varied.plot(x = 'Neighbors', y = ['Guessed', 'False Positive', 'Missed','Likely Accurate'], style = 'o-', title = 'Neighbors vs. Boxes Found')
ax.set_xlabel('Neighbors')
ax.set_ylabel('Number of Boxes')
plt.show()
```
